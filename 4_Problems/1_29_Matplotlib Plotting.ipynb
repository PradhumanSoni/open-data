```python
!pip install datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.head(3)
```

## 🟩 Job Country Bar Chart (1.29.1)

```python
country_count=df.job_country.value_counts(ascending=True)
from matplotlib import pyplot as plt
plt.barh(country_count.index,country_count)
```

---

## 🟨 Average Salary Horizontal Bar Chart (1.29.2)

```python
avg_salary_jobs=df.groupby('job_title_short')['salary_year_avg'].mean()
avg_salary_jobs
plt.barh(avg_salary_jobs.index,avg_salary_jobs)
```

---

## 🟥 Average Salary Trend Line Plot (1.29.3)

```python
avg_salary_jobs=df.groupby('job_title_short')['salary_year_avg'].mean()
avg_salary_jobs
plt.plot(avg_salary_jobs.index,avg_salary_jobs)
```

```python
df['month']=df['job_posted_date'].dt.month
avg_sal_per_month=df.groupby('month')['salary_year_avg'].mean()
plt.plot(avg_sal_per_month.index,avg_sal_per_month)
```
