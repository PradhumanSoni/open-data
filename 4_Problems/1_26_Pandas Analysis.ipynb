```python
!pip install datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df.head()
```

---

## 🟩 Calculate Mean Salary (1.26.1)

```python
print(df['salary_year_avg'].mean())
```

---

## 🟩 Calculate Median Salary (1.26.2)

```python
print(df['salary_hour_avg'].median())
```

---

## 🟨 Average Hourly Salary by Job Title (1.26.3)

```python
df.groupby('job_title_short')['salary_hour_avg'].mean()
```

---

## 🟨 Job with Lowest Salary (1.26.4)

```python
df.groupby('job_title_short')['salary_year_avg'].min().sort_values()
```

---

## 🟨 Job Count by Country (1.26.5)

```python
df.groupby('job_country')['job_title_short'].count()['India']
```
