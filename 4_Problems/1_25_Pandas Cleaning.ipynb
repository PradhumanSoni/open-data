```python
!pip install datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

---

## 🟩 Sort by Salary (1.25.1)

```python
df.dropna(subset='salary_year_avg',inplace=True)
df.sort_values('salary_year_avg')
```

---

## 🟩 Create Remote Column (1.25.2)

```python
df.job_posted_date=pd.to_datetime(df.job_posted_date)

df['Month_Name']=df.job_posted_date.dt.month_name()
df
```

---

## 🟩 Extract Year from Date (1.25.3)

```python
df.job_posted_date.dt.year
```

---

## 🟨 Drop Missing Salaries (1.25.4)

```python
df.drop(labels='salary_hour_avg',axis=1,inplace=True)
df.dropna(subset='salary_year_avg',inplace=True)
df
```
